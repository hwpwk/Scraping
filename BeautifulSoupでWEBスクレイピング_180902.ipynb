```python
from bs4 import BeautifulSoup
import requests
```

```python
html = requests.get('https://www.python.org')
```

```python
# 取得したデータを確認
print(html.text)
```

```python
##############################
```

```python
soup =BeautifulSoup(html.text)
```

```python
# lxmlを勧められているので使用する
soup = BeautifulSoup(html.text, 'lxml')
```

```python
# すべてのtitleタグの中身を取得
titles = soup.find_all('title')
print(titles)
```

```python
# 文字部分(titleの中身)を表示
print(titles[0].text)
```

```python
intro = soup.find_all('div', {'class': 'introduction'})
print(intro)
```

```python
# 文字部分(titleの中身)を表示
print(intro[0].text)
```

```python
html2 = requests.get('http://www.goal.com/jp')
```

```python
soup = BeautifulSoup(html2.text, 'lxml')
```

```python
titles = soup.find_all('title')
print(titles)
```

```python
# titleタグの内容を取りたい場合は、soup.titleでも取得可能
print(soup.title)
```

```python
print(titles[0].text)
```

```python
# こちらでも可能
print(soup.title.string)
```

```python
# aタグを取得
# <A>タグはアンカー（Anchor）の略で、リンクの出発点と到達点を指定するタグ
print(soup.a)
```

```python
soup.find_all('a')
```

```python
# 取得したaタグのHTMLを含まない部分のみ表示
# HTMLを含まない中身をとるものは、stringを利用

list1 = soup.find_all('a')

for i in list1:
    print(i.string)
```

```python
# hrefはリンク先を指定
# href="#"の意味は、「ページのトップに遷移」
soup.a.get('href')
```

```python
for link in list1:
    print(link.get('href'))
```
