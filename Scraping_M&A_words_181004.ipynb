```python
import csv
import requests
from bs4 import BeautifulSoup
```

```python
res = requests.get('http://www.amidaspartners.com/glossary/')
```

```python
# res.contentからHTMLを取り出して、BeautifulSoupで扱えるようにパースします
BeautifulSoup(res.content, "html.parser")
# M&A用語は<div class="word">, class="word"にある
```

```python
# Responseオブジェクトのraise_for_statusメソッド
# →Responseオブジェクトが持つステータスコードが200番台以外だったら、例外を起こす、つまりエラーメッセージを吐き出してスクリプトを停止します。
res.raise_for_status()
soup = BeautifulSoup(res.text, "html.parser")
# 特定のクラスを持つ要素を取得するセレクタは
# [.クラス名]
# というように、ピリオドに続けてクラスを記述
elems = soup.select('.word')
for elem in elems:
    # print(elem)なら<div class="word">偶発債務</div></a>といったタグも含まれてしまうので.stringをつける
    print(elem.string)
```

```python
# windows環境でcsv.writerowメソッドを使う場合ファイルを開く際のopen関数に引数としてnewlineというパラメータに空文字を渡す必要あり
# この引数を追加しないとcsvファイルが一行空きになって書き込まれてしまう
# 参考：https://tonari-it.com/python-csv-writer-writerow/
with open('M&A_word1.csv', 'wt', newline='') as f:
    # csvファイルの書き込み
    writer = csv.writer(f)
    for elem in elems:
        writer.writerow(elem)
# １行だけ書き込むときにはwriterow、複数行を書き込むときにはwriterows
```
